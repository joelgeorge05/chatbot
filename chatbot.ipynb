```python
cardata=pd.read_csv("cardata.csv")
cardata
```

```python
cardata = cardata.drop(cardata.index[100:201])
```

```python
import gradio as gr
```

```python
# Define chatbot function
def answer_question(message, history):
    msg = message.lower()

    if cardata.empty:
        return "Dataset is not loaded."

    if "highest selling price" in msg:
        top = cardata.loc[cardata['Selling_Price'].idxmax()]
        return f"The highest selling price car is {top['Car_Name']} at ₹{top['Selling_Price']} lakhs."

    elif "lowest selling price" in msg:
        top = cardata.loc[cardata['Selling_Price'].idxmin()]
        return f"The lowest selling price car is {top['Car_Name']} at ₹{top['Selling_Price']} lakhs."

    elif "highest present price" in msg:
        top = cardata.loc[cardata['Present_Price'].idxmax()]
        return f"The highest present price car is {top['Car_Name']} at ₹{top['Present_Price']} lakhs."

    elif "lowest present price" in msg:
        top = cardata.loc[cardata['Present_Price'].idxmin()]
        return f"The highest present price car is {top['Car_Name']} at ₹{top['Present_Price']} lakhs."
    else:
        return "Sorry, I couldn't understand your question. You may ask questions related to car details such as fuel type, transmission, ownership history, price range, kilometers driven, and manufacturing year. "
```

```python
# Launch the chatbot interface using updated Gradio format
chatbot = gr.ChatInterface(fn=answer_question, title="Car ChatBot", chatbot=gr.Chatbot(type="messages"))
chatbot.launch(True)
```
